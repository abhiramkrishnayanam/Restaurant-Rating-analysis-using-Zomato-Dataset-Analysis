```python
## Function to remove outliers using the IQR method
def remove_outliers(c_dist, columns):
    for col in columns:
        Q1 = c_dist[col].quantile(0.25)  # First quartile
        Q3 = c_dist[col].quantile(0.75)  # Third quartile
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR  # Lower bound
        upper_bound = Q3 + 1.5 * IQR  # Upper bound
        
        # Filter the data to keep only non-outliers
        c_dist = c_dist[(c_dist[col] >= lower_bound) & (c_dist[col] <= upper_bound)]
    return c_dist

# Remove outliers from the 'count' column
cleaned_c_dist = remove_outliers(c_dist, ['count'])

# Print the shape before and after removing outliers
print(f"Cleaned dataset shape: {cleaned_c_dist.shape}")
```

## Finding the Correlation

```python
numerical_columns=df.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## <font color="blue">Interesting Correlations: </font>

Price range and Aggregate rating (0.44): Moderate positive correlation suggests higher price range may be associated with better ratings.
Votes and Aggregate rating (0.31): Indicates restaurants with more votes tend to have higher ratings, though the relationship is weak to moderate.
Country Code and Longitude (-0.70): Strong negative correlation suggests that certain countries are geographically clustered in terms of longitude.

## <font color="blue">Weak Correlations:</font>

Many variables (e.g., Latitude and Aggregate rating, Average Cost for two and Votes) have near-zero correlations, indicating little to no linear relationship.

## Longitude and Aggregate rating: The correlation coefficient is -0.12.

This is a weak negative correlation, indicating almost no relationship between longitude and the restaurant's rating.

## Latitude and Aggregate rating: The correlation coefficient is 0.00.

This indicates no correlation between latitude and the restaurant's rating.

##  <font color='green'>Conclusion:</font>
There is no significant correlation between a restaurant's location (longitude and latitude) and its rating. The restaurant's geographic position does not appear to influence its aggregate rating based on this dataset.
